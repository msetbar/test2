```python
import pandas as pd
import ast

data_df = pd.read_excel("./non0butlowprecision.xlsx")
src_df = pd.read_csv("./3900_v3_ccmk.csv")
```

```python
src_df = src_df.drop_duplicates(subset=['ArticleNumber', 'Title'])
numtotitle = dict(zip(src_df['ArticleNumber'], src_df['Title']))
```

```python
def extract_unique_filenames(text):
    try:
        # Safely evaluate the literal representation of the list using ast
        file_tuples = ast.literal_eval(text)
        
        # Extract filenames from the list of tuples
        filenames = set(numtotitle[int(filename)] for content, filename in file_tuples)
        
        return list(filenames)
    except (ValueError, SyntaxError):
        return None

# Apply the function to the column and create a new column with unique filenames
data_df['UniqueFilenames'] = data_df['init_chunks'].apply(extract_unique_filenames)

# Display the resulting DataFrame
data_df.head(2)
```
